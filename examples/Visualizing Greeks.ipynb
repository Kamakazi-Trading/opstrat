### Import libraries

```python
import opstrat as op
import pandas as pd
import matplotlib.pyplot as plt
```

### TIME

```python
K=100
St=95
r=4
vs=[10,25,40,60]
ts=[i for i in range(1,101)]
type='c'

df_list=[]
for v in vs:
    vals=[]
    vals_int=[]
    vals_time=[]
    deltas=[]
    gammas=[]
    thetas=[]
    rhos=[]
    vegas=[]
    for t in ts:
        params=op.black_scholes(t=t,K=K,St=St, r=r,v=v, type=type)

        vals.append(params['value']['option value'])
        vals_int.append(params['value']['intrinsic value'])
        vals_time.append(params['value']['time value'])
        deltas.append(params['greeks']['delta'])
        gammas.append(params['greeks']['gamma'])
        thetas.append(params['greeks']['theta'])
        rhos.append(params['greeks']['rho'])
        vegas.append(params['greeks']['vega'])
    df=pd.DataFrame({'vals':vals,'vals_int':vals_int,'vals_time': vals_time,
                     'delta':deltas,'gamma':gammas,'theta':thetas,
                     'rho':rhos,'vega':vegas})
    df_list.append(df)
```

```python
for greek in ['delta','gamma','theta','rho','vega']:
    plt.figure(figsize=(10,6))
    for df,v in zip(df_list,vs):
        plt.plot(ts,df[greek], label=f'{greek} V:{v}%')
    plt.gca().invert_xaxis()
    plt.legend()
    plt.show()
```

```python
greeks=[deltas,gammas,thetas,rhos,vegas]
labels=['delta','gamma','theta','rho','vega']

for greek,label in zip(greeks,labels):
    plt.figure(figsize=(10,6))
    plt.plot(ts,greek, label=label)
    plt.gca().invert_xaxis()
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(ts,vals, label='option value')
plt.plot(ts,vals_time, label='time value')
plt.plot(ts, vals_int, label='intrinsic value')
plt.gca().invert_xaxis()
plt.legend()
plt.show()
```
